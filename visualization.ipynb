```python
import os
import pandas as pd
from matplotlib import cm 
import matplotlib.pyplot as plt
import numpy
```

```python
files= os.listdir('database/')
files
```

```python
scdf=pd.read_csv('database/shopclues_mobiles.csv')
scdf.head()
```

```python
scdf.info()
```

```python
col_to_remove =scdf.columns.str.contains('unnamed',case = False)
```

```python
scdf.drop(scdf.columns[col_to_remove],axis = 1,inplace=True)
```

```python
def display_bar_plot(df,column):
    plt.figure(figsize=(10,5))
    cmap=cm.get_cmap('Spectral')
    plt.title(f'{column} comparision')
    ax=df[column].plot(xticks=df.index,rot=90,legend=True,color='r')
    ax=df[column].plot(kind='bar',xticks=df.index,rot=90,colormap=cmap)
    ax.set_xticklabels(df.name)
    plt.ylabel(f'{column} range')
    plt.show()
```

```python
def compare_price(df,file):
    cmap=cm.get_cmap('tab10')
    ax=df.plot(y=['price','sale_price','Striked Price'],kind='bar',xticks=df.index,rot=90,colormap=cmap,figsize=(15,5))
    ax.set_xticklabels(df.name)
    plt.ylabel(f' range')
    plt.title(f'comparision')
    fig = plt.gcf()
    plt.show()
    fig.savefig(f'{file}.png')
```

```python
def compare_review_rating(df,file):
    cmap=cm.get_cmap('copper')
    ax=df.plot(y=['rating','reviews'],kind='bar',xticks=df.index,rot=90,colormap=cmap,figsize=(15,5))
    ax.set_xticklabels(df.name)
    plt.ylabel(f'range')
    plt.title(f'comparision')
    fig = plt.gcf()
    plt.show()
    fig.savefig(f'{file}.png')
```

```python
compare_review_rating(df,'shopclues_review')
```

```python
display_bar_plot(scdf,'sale_price')
```

```python
compare_price(df,'shopclues_prices_stats')
```
